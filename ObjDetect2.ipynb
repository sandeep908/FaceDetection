```python
import cv2
import matplotlib.pyplot as plt
```

```python
image= cv2.imread("/content/drive/MyDrive/Colab Notebooks/face1.jpg")
# plt.imshow(image)
image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
xml_data = cv2.CascadeClassifier("/content/drive/MyDrive/Colab Notebooks/haarcascade_frontalface.xml")
detecting = xml_data.detectMultiScale(image_grey, minSize = (30, 30))
amountDetecting = len(detecting)

if amountDetecting !=0:
  for(a, b, width, height) in detecting:
    cv2.rectangle(image_rgb, (a,b),(a+height, b+width),(0,275,0),9)

plt.subplot(1,1,1)
plt.imshow(image_rgb)
plt.show()
```
